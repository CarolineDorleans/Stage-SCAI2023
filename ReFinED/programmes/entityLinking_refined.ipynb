```python
import json

import glob
from refined.inference.processor import Refined


refined = Refined.from_pretrained(model_name='wikipedia_model_with_numbers',
                                  entity_set="wikipedia")


def lire_fichier(chemin):
    with open (chemin, "r", encoding="utf-8") as fichier:
        texte=fichier.read()
        return texte
def stocker(chemin, contenu):
    w =open(chemin, "w")
    w.write(json.dumps(contenu , indent = 2))
    w.close()

#path_corpora = "../DATA/ELTeC-Fra_2023/*/*/*/*.txt"
path_corpora = "../DATA/ELTeC-Fra_2023/ADAM/ADAM_krakenbase/REF/*.txt"


for path in glob.glob(path_corpora):
    print("Entrée : ",path)
    txt=lire_fichier(path)
    spans = refined.process_text(txt)
    ide= 1
    dico = {}
    dico_sent_tok={}
    for span in spans:
        print ("___________________________________________________")
        dico = span.__dict__ 
        for cle, valeur in dico.items() :
            candidate_entities = dico.get("candidate_entities")
            coarse_mention_type = dico.get("coarse_mention_type")
            coarse_type = dico.get("coarse_type")
            date = dico.get("date")
            description_scores = dico.get("description_scores")
            doc_id = dico.get("doc_id")
            entity_linking_model_confiance_score = dico.get("entity_linking_model_confiance_score")
            failed_class_check = dico.get("failed_class_check")
            gold_entity = dico.get("gold_entity")
            predicted_entity = dico.get("predicted_entity")
            predicted_entity_types = dico.get("predicted_entity_types")
            pruned_candidates = dico.get("pruned_candidates")
            start = dico.get("start")
            text = dico.get("text")
            top_k_predicted_entities = dico.get("top_k_predicted_entities")
            
            dico_sent_tok[ide]={}
            dico_sent_tok[ide]["candidate_entities"]=candidate_entities
            dico_sent_tok[ide]["coarse_mention_type"]=coarse_mention_type
            dico_sent_tok[ide]["coarse_type"]=coarse_type
            dico_sent_tok[ide]["date"]=date
            dico_sent_tok[ide]["description_scores"]=description_scores
            dico_sent_tok[ide]["doc_id"]=doc_id      
            dico_sent_tok[ide]["top_k_predicted_entities"]=top_k_predicted_entities
            dico_sent_tok[ide]["text"]=text
            dico_sent_tok[ide]["start"]=start
            dico_sent_tok[ide]["pruned_candidates"]=pruned_candidates
            dico_sent_tok[ide]["predicted_entity_types"]=predicted_entity_types
            dico_sent_tok[ide]["predicted_entity"]=predicted_entity
            dico_sent_tok[ide]["gold_entity"]=gold_entity
            dico_sent_tok[ide]["failed_class_check"]=failed_class_check
            dico_sent_tok[ide]["entity_linking_model_confiance_score"]=entity_linking_model_confiance_score
            ide=+1
        stocker("%s_entitylinker-refined-fr.json"%(path),dico_sent_tok)
```
